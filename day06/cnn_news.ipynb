BeautifulSoup 으로 불러오기

```python
import requests
from bs4 import BeautifulSoup

url = "http://edition.cnn.com/2025/12/21/asia/philippines-christmas-decor-traditions-intl-hnk-dst"
headers = {
  "User-Agent" : "Moziila/5.0" 
}

res = requests.get(url, headers=headers, timeout=10)
res.raise_for_status()

soup = BeautifulSoup(res.text, "html.parser")

article = soup.find("article")
paras = article.find_all("p") if article else soup.find_all("p")

text = "\n".join(p.get_text(" ", strip=True) for p in paras)
text[:500]
```

텍스트로 붙혀서 사용해보기

```python
text = """
Via Arboleda stepped back to admire the six-foot tree glittering in her living room.

Though the narra trees along Manila’s sidewalks had yet to shed their leaves, she was among many Filipinos already unpacking their Christmas decorations in eager anticipation of December 25.

In September, colorful ornaments began appearing along her commute, lifting her spirits beneath the dark sky of the rainy season.

“We believe the bigger the celebration, the better,” Arboleda, 27, an advertising professional from Manila, told CNN.

The Philippines has the longest Christmas period in the world, with festivities spanning from September to January.
""".strip()
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
```

기사 안 10줄 보기

```python
#문장 단위
sents = sent_tokenize(text)
sents[:10]
print(sents)
```

문장 앞 5줄 보기

```python
nltk.download('punkt_tab') #nltk 최신버전일 때 추가
from nltk.tokenize import sent_tokenize
# 문장 단위
sents = sent_tokenize(text)
print(sents)
```

```python
# 단어 단위
tokens = word_tokenize(text)
print(tokens[:100])
```

```python
from nltk.tokenize import RegexpTokenizer
regToken = RegexpTokenizer('[\w]+') #문자 한자 이상을 분석 단위로 잡겠다.capitalize
regToken.tokenize(text[:100])
```

```python
from nltk.tag import pos_tag
nltk.download('averaged_perceptron_tagger_eng')

sentence=text[:1000]
# sentence # 앞에 문자 1000자만 나와
tagged_list = pos_tag(regToken.tokenize(sentence)) #regToken 정규표현식 하면 문자만 가지고 와
tagged_list #튜플로 받은 거임 튜플 0에는 단어, 1에는 품사
```

```python
noun_list=[t[0] for t in tagged_list if t[1]=='NNP'] 
noun_list
```

```python
noun_list=[t[0] if t[1]=='NN' else '-' for t in tagged_list]

#명사만 추출하고 싶으면
noun_list
```
