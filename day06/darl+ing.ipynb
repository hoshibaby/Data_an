```python
nltk.download('punkt_tab') #nltk 최신버전일 때 추가
from nltk.tokenize import sent_tokenize
# 문장 단위
sents = sent_tokenize(text)
print(sents)
```

단어 단위로 토큰을 만들자.

```python
tokens = word_tokenize(text)
print(tokens[:200])
```

정규식을 만들어보자.

```python
from nltk.tokenize import RegexpTokenizer
regToken = RegexpTokenizer('[\w]+') #문자 한자 이상을 분석 단위로
regToken.tokenize(text[:200])
```

태깅해보자!!!

```python
from nltk.tag import pos_tag
nltk.download('averaged_perceptron_tagger_eng')

sentence = text[:2000]

tagged_list = pos_tag(regToken.tokenize(sentence))
tagged_list
```

t는 “(단어, 품사태그)” 한 쌍(tuple)
t[0] = 단어(word)
t[1] = 품사 태그(tag)
그래서 t[1] == 'NNP'는 고유명사인지 검사하는 조건

```python
noun_list = [t[0] for t in tagged_list if t[1]=='NNP']
noun_list
```

명사만 추출하고 싶으면

```python
noun_list=[t[0] if t[1]=='NN' else '-' for t in tagged_list]
noun_list
```

반복되는 단어의 빈도를 이미지로

```python
from nltk import Text

text=Text(regToken.tokenize(text))
print(text)
```

출현빈도가 가장 높은 단어 10개로 차트 작성

```python
import matplotlib.pyplot as plt
text.plot(10)
```
