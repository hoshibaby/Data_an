회귀분석 전 라이브러리 설치된거 확인함. 라이브러리 가지고 오자.

```python
from scipy import stats
from statsmodels.formula.api import ols, glm
```

ols
정답과 결과치 오차를 제곱하여 다 더해서 나눠 평균을 내.
최소제곱!!!!!!!!!

퀄리티 평균차가 유의미한지 기각될 값인지 비교 -> 검정작업부터하자

```python
red_wine_quality=wine.loc[wine['type']=='red', 'quality'] #레드 행만 다 골라라!
red_wine_quality
```

하나의 행만 가지고 와서 시리즈 형태로 나왔어.

```python
white_wine_quality=wine.loc[wine['type']=='white', 'quality']
white_wine_quality
```

t테스트의 결과가 아래와 같이 나옴.
pvalue -24승이니까 0.0000 이고, 0.05보다는 작다. 유의미다. 귀무가설 채택한다.

```python
stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var=False)
```

데이터 모델링
t-검정과 회귀분석으로 그룹 비교하기

y=a1x1+a2x2+ .... +b의 수식을 아래와 같이 쓴거야.

```python
Rformula='quality~fixed_acidity+volatile_acidity+citric_acid+residual_sugar+chlorides+free_sulfur_dioxide+total_sulfur_dioxide+density+pH+sulphates+alcohol'
regression_result=ols(Rformula, data=wine).fit() #regression 이건 회귀분석, Rformula 선형회귀 수식, fit은 훈련
```

regression 이건 회귀분석, Rformula 선형회귀 수식, fit은 훈련

Dep. Variable: 종속변수는 quality이고, Model은 OLS 를 쓴다. 회귀분석 함수는 Least Squares 를 쓴다.
시간 나오고
No. Observations: 데이터 수
Covariance Type:	nonrobust	아래들은 독립변수들임. 와인퀄리티에 어떤 영향을 주는가 표현을 함.
density 가 큰 영향을 줘. 
lntercept는 바이어스 상수값.???????????????????
