```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

데이터 프레임 만들기
시간이 증가하면 시험 점수도 증가하게 데이터를 만들었어.


```python
x=[1,2,3,4,5]
y=[50, 55, 65, 70, 80] 
data=pd.DataFrame({'x':x, 'y':y})
data
```

상관관계가 0.8 이상이니까 엄청 높아.

```python
corr1=data.corr(method='pearson')
corr1
```

차트로 만들어보자
시간이 증가하면 시험 점수도 증가하게 데이터를 만들었기 때문에 상관이 높다.

```python
plt.figure(figsize=(12,10))
sns.relplot(data=data, x='x', y='y')
```

```python
x1=np.random.rand(100)
y1=np.random.rand(100) #랜덤하게 0~1 사이의 수를 만들어써.
data1=pd.DataFrame({'x':x1, 'y':y1}) #특별한 규칙없이 데이터 프레임 만들었음.
data1
```

```python
corr2=data1.corr() #피어슨 안적어도 연속된 수는 피어슨 적용
corr2
```

-0.07 0.02 이하는 상관이 없다고 해석을 해.
차트 그려보자.

```python
sns.relplot(data=data1, x='x', y='y')
```
