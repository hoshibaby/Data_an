데이터 읽기

```python

data=pd.read_excel('data/04_data1.xlsx')
data.head()
data2020_good=data['2020_good']
data2020_bad=data['2020_bad']
data2020_common=data['2020_common']
data2020_good
data2020_bad
```

```python
plt.figure(figsize=(20,4))
plt.plot(data2020_good, color='b', marker='s', label='good')
plt.plot(data2020_bad, color='r', marker='o', label='bad')
plt.plot(data2020_common, color='g', marker='^', label='common')
plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2020 fine dust(pm10) Good Line Graph')
plt.ylim(0,350)
plt.legend()
plt.grid()
plt.show()
```

연도별 미세먼지 좋음 상태 그래프 그리기

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 4))

for year in range(2015, 2021):
  year = str(year)+'_good'
  chartdata = data[year]
  plt.plot(chartdata, marker='o', label=year)

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015~2020 Fine Dust(pm10) Good Line Graph' )
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20, 4))

colors = ['tab:blue', 'tab:orange', 'tab:green',
          'tab:red', 'tab:purple', 'tab:brown']

for i, year in  enumerate(range(2015, 2021)):
  col = str(year)+'_good'
  chartdata = data[col]
  chartdata.plot(kind='bar', label=col, color=colors[i], alpha=0.7)

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015~2020 Fine Dust(pm10) Good Bar Graph' )
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(20, 4))

for year in range(2015, 2021):
  col = str(year)+'_good'
  chartdata = data[col]
  plt.bar(chartdata.index, chartdata.values, label=col)

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.title('2015~2020 Fine Dust(pm10) Good Line Graph' )
plt.legend()
plt.grid()
plt.show()
```

2015~2020 막대 그래프 그리기

```python
import matplotlib.font_manager as fm
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False #- 있으면 깨져 빼자.
```

```python
plt.figure(figsize = (10,4))

index = np.arange(7)
data2 = data1.loc['Seoul':'Ulsan', '2015_good':'2020_good'] 
for year in range(2015, 2021):
  year = str(year)+'_good' #숫자랑 문자랑 결합하면 에러나서 str 붙혀서 문자열 만들고 합쳐
  chartdata = data2[year] # data2에서 year 가지고 오는 거
  plt.bar(index, chartdata, width=0.15, label=year)
  index = index + 0.15

plt.xlabel('Area')
plt.ylabel('Micrometer')
plt.xticks(index-0.5, ['서울', '부산', '대구', '인천', '광주', '대전', '울산'])
plt.title('2015~2020 미세먼지 농도')
plt.legend()
plt.show()
```

수평 막대그래프를 누적 그래프로 그려보자. 2017~2020 보통날

```python
plt.figure(figsize = (10,4))

index = np.arange(7)
data2 = data1.loc['Seoul':'Ulsan', '2017_common':'2020_common'] 
for year in range(2017, 2021):
  year = str(year)+'_common' #숫자랑 문자랑 결합하면 에러나서 str 붙혀서 문자열 만들고 합쳐
  chartdata = data2[year] # data2에서 year 가지고 오는 거
  plt.barh(index, chartdata, label=year)
plt.legend()
plt.xlim(150, 300)
plt.yticks(index, ['서울', '부산', '대구', '인천', '광주', '대전', '울산'])
```

```python
iris_df=pd.read_csv('data/iris.csv')
iris_df
```

```python
petalLength=iris_df['petal_length']
petalWidth=iris_df['petal_width']

plt.figure(figsize=(10,4))
plt.scatter(petalLength, petalWidth)
plt.grid()
plt.show()
```

```python
petalLength=iris_df['petal_length']
petalWidth=iris_df['petal_width']
sepalLength=iris_df['sepal_length']
sepalWidth=iris_df['sepal_width']

plt.figure(figsize=(10,4))
plt.scatter(petalLength, petalWidth, label='꽃잎')
plt.scatter(sepalLength, sepalWidth, label='꽃받침')
plt.legend()
plt.xlabel('length')
plt.ylabel('width')
plt.grid()
plt.show()
```

히스토그램 그래프 그려보자

꽃잎의 길이가 어떻게 분포되어있나.

```python
setosa = iris_df.loc[iris_df.species=='setosa', ['species', 'petal_length']] #loc는 행과 열을 선택한다는 뜻
versicolor=iris_df.loc[iris_df.species=='versicolor',['species','petal_length']]
verginica=iris_df.loc[iris_df.species=='virginica',['species','petal_length']]

setosa
versicolor
verginica
```

```python
plt.Figure(figsize=(10,6))
plt.hist(setosa['petal_length'], bins=8, label='setpsa Legth', alpha=0.5 ) 
plt.hist(versicolor['petal_length'], bins=5, label='versicolor Length', alpha=0.5 )
plt.hist(verginica['petal_length'], bins=5, label='verginica Length', alpha=0.5 )
#내가 구간bin을 안잡으면 10개로 그려
```
