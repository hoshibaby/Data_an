```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup as bs
import urllib.request as req
```

```python
key='1413931c9f3333c9669e0f37d6067bed0272be5914ae7256c959874092fc2fcf'
lineno=23

url="http://apis.data.go.kr/6260000/BusanBIMS/busInfo"
params='?ServiceKey='+key+'&lineno='+str(lineno)
savepath='../data/businfo.xml' #공공db나 외부 자료를 계속 부르면 트레픽 느니까 저장을 해버려.
req.urlretrieve(url+params, savepath) #(src, des) 노선 정보 조회 페이지에 들어가서 저장
```

```python
xml=open(savepath, 'r', encoding='utf-8').read()
soup=bs(xml, 'xml')
# soup
lineId=None
for item in soup.find_all('item'):
  busno=item.find('buslinenum').text #text 붙혀주기 전에는 이것저것 다나와
  if busno !=str(lineno):
    lineId=item.find('lineid').text
    break
print(busno, lineId)  #라인아이디 나왔다
```
