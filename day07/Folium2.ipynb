지리정보 분석 후 맵 생성

```python
import numpy as np
import pandas as pd
import folium
import requests
import json
```

```python
url='https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'
r=requests.get(url)
c=r.content
seoul_geo = json.loads(c)
seoul_geo
```

서울 시청 좌표를 가지고 오자.



```python
lat = 37.56682645618825
lon = 126.97865791770842
map=folium.Map(location=[lat, lon], zoom_start=11)
folium.GeoJson(
  seoul_geo,
  name='지역구'
).add_to(map)
map
```

구별로 goupby 하면 서울시 구별 커피빈 매장 갯수를 확인하고 색칠해보자.

파일을 열어보자. 파일이 와장창 깨져있어. cp949로 가지고 와야해. index가 저장된 파일 가지고 오게되면 index_col=0으로 0번째 줄에
인덱스를 설정해줘.


```python
cd=pd.read_csv('../data4/CoffeeBean.csv', encoding='cp949', index_col=0)
cd
```

나는 주소와 구가 필요해.

```python
addrs=cd['address'].tolist()
# addrs
addr_list=[]
for addr in addrs:
  # print(addr.split()) #공백을 기준으로 나눠줘
  addr_list.append(addr.split()[:2]) #0, 1만 가지고 와
addr_list
```

데이터 프레임을 만들어서 groupby 해보려고 프레임을 만들었어.

```python
df=pd.DataFrame(addr_list, columns=['sido', 'gugun'])
df
```

전국 데이터가 다 있고, 서울이라는 이름이 3개가 잇어. 똑같은 이름으로 as로 만들어줘야해.

```python
df['sido'].unique()
```

이름을 엘리어스해서 고쳐주자!!!

```python
addr_alias={'서울':'서울특별시', '서울시':'서울특별시', '부산시':'부산광역시', '부산':'부산광역시'}
df['sido']=df['sido'].replace(addr_alias)
df['sido'].unique()
```

서울데이터만 가지고 와보자!
df1은 서울데이터만 가지고 있는 것도 확인.

```python
df1=df.loc[df['sido']=='서울특별시',:]
df1['sido'].unique()
```

그룹바이해서 구별 카운팅을 해보자.

```python
df_grouped=df1.groupby('gugun')['sido'].count()
df_grouped
```

이제 경계선 좌표에 집어넣어보자.
geo data에 properties가 있고, 거기 name 기준으로 맞춘다.

```python
folium.Choropleth(
  geo_data=seoul_geo,
  data=df_grouped, #구 이름이 있어.
  name='choropleth',
  key_on='properties.name', 
  fill_color='YlOrRd', #이 순서대로 높아짐
  fill_opacity=0.5,
  line_opacity=0.5, #채우는 색의 불투명도
  legend_name='구별 커피빈 수'
).add_to(map)
map
```

우측 상단에 48개의 빈을 균등하게 나눠서 색을 칠했어. 좀 구분이 잘 안되는 것 같아 bin 수정
bind을 조절하면 다양하게 색깔을 표현할 수 있는게 이 것의 장점.

```python
folium.Choropleth(
  geo_data=seoul_geo,
  data=df_grouped, #구 이름이 있어.
  name='choropleth',
  key_on='properties.name', 
  fill_color='YlOrRd', #이 순서대로 높아짐
  fill_opacity=0.5,
  bins=[0, 2, 5, 15, 20, 25, 50], #데이터가 조밀한 곳은 세세하게 나누고, 듬성한 곳은 하나의 빈으로 나오고
  line_opacity=0.5, #채우는 색의 불투명도
  legend_name='구별 커피빈 수'
).add_to(map)
map
```
