마커가 너무 많으면 어지러우니까아아..
확대했을때는 마크가 다 보이게, 축소했을때는 마크가 몇개 있습니다 표시

```python
from folium import plugins
```

```python

N = 100 
data = np.array( 
[
np.random.uniform(low=35, high=38, size=N), # 랜덤으로 좌표를 만들어라 
np.random.uniform(low=125, high=129, size=N), # 랜덤으로 좌표를 만들어라
] ).T #행과 열을 바꿔 위도 경도 한쌍이 되게, 100개의 위도경도 쌍이됨. 2차원 배열 만들었어.
popups = [str(i) for i in range(N)] # 0부터 99까지 숫자를 문자열로 만들어서 popup으로 쓰겠다.
m1 = folium.Map([lat, lon], zoom_start=8) 
plugins.MarkerCluster(data, popups=popups).add_to(m1)  #축소하면 이 영역에 마크가 몇개 있습니다.
# m1.save(os.path.join('results', 'Plugins_1.html')) 
m1
```

실제 데이터도 해보자.
nan도 있고, 카메라 대수가 0 인 것도 있고.


```python
df=pd.read_csv('../data4/CCTV_20190917.csv')
df
```

부산 진구 cctv설치 지역 자료야
카메라 대수 0인거 제외하자.

```python
df1=df.loc[df['카메라대수']>=1, :]
df1
```

```python
# df1['소재지도로명주소'] = df1['소재지도로명주소'].apply(str.replace(' ', '')) 이거 하니 안나와서 지움
# apply는 각 필드에 여러함수를 적용한다는 뜻
popups=df1['소재지도로명주소'].tolist() #시리즈를 리스트로 만들어줘
popups
```

전체 지도 좌표에 중심좌표. 위도 한가운데 경도 한가운데 잡으면 중심좌표 나오겠지?

```python
lat_avg=np.array(df1['위도'].tolist()).mean()
lat_avg 
```

```python
lon_avg=np.array(df1['경도'].tolist()).mean() #시리즈를 tolist로 만들고 연산
lon_avg 
```

지도를 만들자.

```python
m=folium.Map(location=[lat_avg, lon_avg], zoom_start=15) #부산진구의 좌표만큼 만 잡았어.
m
```

```python
data=np.array([df1['위도'].to_list(), df1['경도'].to_list()]).T
for i in range(len(data)):
  folium.Marker([data[i][0], data[i][1]], popup=popups[i],
                icon=folium.Icon(color='red', icon='info-sign')).add_to(m)
data
```

```python
import os
m=folium.Map(location=[lat_avg, lon_avg], zoom_start=12)
plugins.MarkerCluster(data, popup=popups).add_to(m)
m.save(os.path.join('../data4','cctvmap.html'))
m
```

클릭한 곳에 마크가 생기게 해보자. p16

```python
m=folium.Map(location=[lat, lon])
m.add_child(folium.ClickForMarker(popup='마커')) #지도에 클릭하면 마커가 생겨
m.add_child(folium.LatLngPopup()) #클릭하니가 위도 경도가 표시가 되는 핀업이 생겨
m
```
